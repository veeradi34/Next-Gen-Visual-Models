```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import tensorflow as tf
import imghdr
```

```python
import kagglehub


path = kagglehub.dataset_download("paultimothymooney/chest-xray-pneumonia")

print("Path to dataset files:", path)
```

```python
train_data_path='/kaggle/input/chest-xray-pneumonia/chest_xray/train'
test_data_path='/kaggle/input/chest-xray-pneumonia/chest_xray/test'
```

```python
train_data=tf.keras.utils.image_dataset_from_directory(train_data_path)
test_data=tf.keras.utils.image_dataset_from_directory(test_data_path)
```

```python
train_size=int(len(train_data)*0.7)+1
val_size=int(len(train_data)*0.2)+1

train=train_data.take(train_size)
val=train_data.skip(train_size).take(val_size)

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Dense,Flatten,Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
model=Sequential([Conv2D(32,(3,3),activation='relu',input_shape=(256,256,3)),MaxPooling2D(2,2),Conv2D(16,(3,3),activation='relu'),MaxPooling2D(),Flatten(),Dense(256,activation='relu'),Dense(1,activation='sigmoid')])
```

```python
model.compile(loss=BinaryCrossentropy(),optimizer=Adam(),metrics=['accuracy'])
```

```python
model.fit(train,epochs=2,validation_data=val)
```

```python
loss,accuracy=model.evaluate(test_data)
```
